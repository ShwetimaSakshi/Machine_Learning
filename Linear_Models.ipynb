Question: Explain the mathematical derivation for training the
linear regression model with ordinary least squares (OLS) solution.

---

(You should use LaTeX commands in a markdown cell to write the math)

Note: Limit your answer to this cell

Answer: The mathematical derivation of the OLS estimator for a simple linear regression model is as follows.
Let's consider a simple linear regression model with one independent variable (predictor) and one dependent variable (response). We assume that the relationship between the independent variable (X) and the dependent variable (Y) is linear and can be represented as:

$$Y = w₀ + w₁X + ε$$

Here:
1. Y is the dependent variable. 
2. X is the independent variable. 
3. w₀ is the intercept (the value of Y when X is 0). 
4. w₁ is the slope (the change in Y for a one-unit change in X). 
5. ε is the error term, representing the random variation or noise in the data.


Now, suppose we have the dataset with n observations: 
$$(X₁, Y₁), (X₂, Y₂), ..., (Xₙ, Yₙ)$$

For any observation at position i, the predicted values Ŷᵢ is calculated using the linear regression model:
$$Ŷᵢ = w₀ + w₁Xᵢ + ε$$

So, for all the n observations, we can write:

$$ Y₁ = w₀ + w₁Xᵢ + ε₁
\\Y₂ = w₀ + w₂X₂ + ε₂
\\⋮
\\Yₙ = w₀ + wₙXₙ + εₙ 
$$


$$\begin{bmatrix}  Y_{1}  \\ Y_{2} \\.\\.\\.\\Y_{n}\end{bmatrix}=\begin{bmatrix} 1 & X_{1} \\ 1 & X_{2} \\.\\.\\.\\1 & X_{n} \end{bmatrix}  \begin{bmatrix} w_{0} \\ w_{1} \end{bmatrix}  + \begin{bmatrix}  ϵ_{1}  \\ ϵ_{2} \\.\\.\\.\\ϵ_{n}\end{bmatrix} $$

or, $$ Y = XW + ε $$

where 
$$ Y = \begin{bmatrix} Y_{1},...., Y_{n} \end{bmatrix} ^{T} $$
$$ ϵ = \begin{bmatrix} ϵ_{1},...., ϵ_{n} \end{bmatrix} ^{T}  $$
$$ W = \begin{bmatrix} W_{0},W_{1} \end{bmatrix} ^{T} $$
$$ X =  \begin{bmatrix} 1 & X_{1}\\.&.\\.&.\\.&. \\ 1 & X_{n} \end{bmatrix} $$

we want to minimize the sum of squared residuals:
$$ RSS(W) = ε^{T}ε $$
$$ =(Y−XW)^{T}(Y−XW) $$
$$ = Y^{T}Y - W^{T}X^{T}Y - Y^{T}X^{T}W + W^{T}X^{T}XW \rightarrow  \min_{ W_{0}, W_{1} } $$

Equating the partial derivative to zero:
$$ \frac{\partial RSS(  \widehat{W}  ) }{\partial  \widehat{W} } = −2X^{T}Y + 2X^{T}X\widehat{W} = 0 $$

Thus, the OLS estimator:
$$ \widehat{W} =(X^{T}X)^{-1}X^{T}Y $$

Answer: The coefficients for each feature for both models are different due to the fundamental differences in the two regression techniques. The reasons for this difference can be:
1. Linear regression is used for predicting a continuous outcome variable (numeric value) while logistic regression is used for binary classification problems where the outcome variable is categorical with two classes (e.g., 0 or 1, Yes or No).
2. The model assumes a linear relationship between the independent variables (features) and the dependent variable while logistic regression models the probability of belonging to one class (e.g., class 1) as a function of the independent variables.
3. Coefficients in linear regression represent the change in the dependent variable for a one-unit change in the corresponding independent variable while holding all other variables constant whereas coefficients in logistic regression represent the log-odds of the probability of the event occurring for a one-unit change in the corresponding independent variable.
4. Coefficients are chosen to minimize the sum of squared differences between the predicted values and the actual values in linear regression while logistic regression uses the logistic or sigmoid function to transform a linear combination of the independent variables into probabilities between 0 and 1.

### Explain the mathematical derivation you used for implementing the training of your linear and logistic regression models.

Answer: 
Formula of logistic function:
$$P = \frac{1}{1+ e^{-( W_{0} +  W_{1}X)} } $$

We all know the equation of the best fit line in linear regression is:
$$  Y = W_{0} +  W_{1}X $$

Instead of y we are taking probabilities (P)
$$  P = W_{0} +  W_{1}X $$

But since the probability value shall be bewteen 1 and 0, hence, we shall take  “odds” of P
$$  \frac{P}{1-P} = W_{0} +  W_{1}X $$

We know that odds can always be positive which means the range will always be (0,+∞ ). The problem here is that the range is restricted and we don’t want a restricted range because if we do so then our correlation will decrease. To control this we take the log of odds which has a range from (-∞,+∞).
$$  \log{\frac{P}{1-P}} = W_{0} +  W_{1}X $$

multiply by exponent on both sides and then solve for P.
$$  exp[\log{\frac{P}{1-P}}] = exp(W_{0} +  W_{1}X) $$

$$  e^{\ln[{\frac{P}{1-P}}]} = e^{W_{0} +  W_{1}X} $$

$$  \frac{P}{1-P} =  e^{W_{0} +  W_{1}X} $$

$$ P =  e^{W_{0} +  W_{1}X} - Pe^{W_{0} +  W_{1}X} $$

$$ P =  P[\frac{e^{W_{0} +  W_{1}X}}{P} - e^{W_{0} +  W_{1}X}] $$

$$ 1 =  \frac{e^{W_{0} +  W_{1}X}}{P} - e^{W_{0} +  W_{1}X} $$

$$ P[1 + e^{W_{0} +  W_{1}X}] =  e^{W_{0} +  W_{1}X} $$

$$ P =  \frac{e^{W_{0} +  W_{1}X}}{1 + e^{W_{0} +  W_{1}X}} $$

$$ P =  \frac{1}{1 + e^{-(W_{0} +  W_{1}X)}} $$

This is our logistic function, also called a sigmoid function.
